# Iris Dataset Notebook
## About the Iris Dataset
The [Iris flower data set](https://en.wikipedia.org/wiki/Iris_flower_data_set) is a dataset containing 50 samples of each of three species of Iris flowers; Setosa, Versicolor and Virginica. Along with the species, the dataset contains measurements of the sepal length, sepal width, petal length, and petal width for each sample. 

This dataset was created by the statistician Ronald Fisher in 1936 as an example of *linear discriminant analysis*, a method used to find a linear combination of features that characterises or separates two or more classes of objects. So the purpose of the dataset is to use measurements on the morphology of the samples and use that to distinguish the species from each other.

The purpose of this notebook is to explain the data set including the difficulty in writing an algorithm to separate the three classes of iris based on the variables in the dataset.

### Package Imports

```python
import pandas as pd                     # For interacting with data sets
import matplotlib.pyplot as pl          # Plotting package
import seaborn as sns                   # For styling plots
```

## Load the Dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv") # Load the dataset
df                                                                                              # Display the dataset
```

Here we can see the dataset in it's entirety. As you can see it is made up of 6 columns. The first column identifies the sample, the following four contain measurements of the sample's size, and the final column tells us the species. Here we have all the data we need, but it isn't very readable for us. In order to properly visualize the dataset we can use seaborn to display it on a number of plots. Documentation of seaborn's plots can be found [here](https://seaborn.pydata.org/examples/index.html).

```python
sns.violinplot(x='species', y='sepal_width', data=df, inner='quartile')
pl.show()
sns.violinplot(x='species', y='sepal_length', data=df, inner='quartile')
pl.show()
sns.violinplot(x='species', y='petal_width', data=df, inner='quartile')
pl.show()
sns.violinplot(x='species', y='petal_length', data=df, inner='quartile')
pl.show()
```

```python
# Plot the Iris data set with a pair plot.
sns.pairplot(df, hue="species")
```

## Difficulty in Differentiating Species
Now that we have something much more readable, we can see that each of the iris species are clustered together with their homogenous samples. We can see that the setosa species are usually linearly separable from the other two species. There is, however a lot of overlap in the dimensions of the versicolor and virginica samples. It is this overlap that makes is difficult even for us to differentiate the species, let alone for an algorithm. The best solution to solving this issue is to use a neural network, and to train it until it can predict the species to a degree of accuracy.
